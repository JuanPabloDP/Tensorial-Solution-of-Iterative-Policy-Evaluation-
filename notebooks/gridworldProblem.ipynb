## Examples of iterations

Here are done by hand the first iterations, which are as the results of the book. Book results are truncated by one decimal.

![First Iterations](../figures/figure5.png){width=600}

### First iteration example

```python
RV=R+V_0
RV
```

```python
RVP=np.tensordot(P,RV,axes=1)
RVP
```

```python
V_1=PI @ RVP 
V_1
```

### Second iteration example

```python
V_2=PI @ np.tensordot(P,R+V_1,axes=1)
V_2
```
